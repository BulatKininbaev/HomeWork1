```python
"""
Пользователь в бесконечном цикле вводит числа через консоль до тех пор, пока не введёт что-то, отличное от числа (после этого цикл завершается). В конце программа должна вывести сумму всех введённых чисел.
"""
may_continue = True
all_sum = 0
while may_continue:
    any_number = input('Введите или символ')
    if not any_number.isnumeric():
        may_continue = False
        continue
    all_sum += int(any_number)
print(f'Общая сумма ={all_sum}')
```

```python
"""
Пользователь вводит строку через консоль (состоящую только из букв в разном регистре). Программа должна создать новую строку, содержащую только те буквы из исходной строки, которые были в верхнем регистре.

"""

any_str = input('Введите строку')
upper_str = ''
for one_char in any_str:
    if one_char.isupper():
        upper_str += one_char
print(upper_str)
```

```python
"""
Классическая игра FizzBuzz. Программа в цикле выводит в консоль числа от 1 до 100.
"""
for i in range(1,100):
    if i%3==0 and i%5==0:
        print('FizzBuzz')
    elif i%3==0 and not i%5==0:
        print('Fizz')
    elif not i%3==0 and i%5==0:
        print('Buzz')
    else:
        print('мимо кассы')
```

```python
"""
Пользователь вводит строку, содержащую любые символы (буквы, цифры и специальные символы). Программа должна вычислить и вывести сумму всех цифр, присутствующих в этой строке.

"""
any_string = input('Введите любую строку цифры и буквы')
summa=0
for char in any_string:
    if char.isnumeric():
        summa+=int(char)
print(summa)
```

```python
"""
Пользователь вводит через консоль 10 чисел. Программа должна определить максимальное число и вывести его в консоль.

"""
i=1
max_number = None
while i<10:
    in_str = input("Введите число")
    if in_str.isnumeric():
        i+=1
        if (max_number is None) or (max_number<int(in_str)):
            max_number=int(in_str)
print(f'Максимальное число из введенных 10-ти = {max_number}')


```

```python
"""
Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах
 Не стал проверять на то что ввели цифры и их количество - как я понял это задача на работу со троками .
"""
numbers = input('Ведите любую строку')
middle_numbers = numbers[1:-1]
print(numbers[0]+middle_numbers[::-1]+numbers[-1])
```

```python
"""
Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.
"""
day_to_vocation=int(input('Введите количество дней до отпуска'))
week_day = day_to_vocation//6+day_to_vocation//7
print(f'количество выходных дней до отпуска {week_day}')
```

```python
"""
Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой
"""
shoklad_size=input('введите через запятую длину и ширину шоколадки и размер куска пример (5 ,9 ,15)')
dlinna=int(shoklad_size.split(',')[0])
shirina=int(shoklad_size.split(',')[1])
kusok=int(shoklad_size.split(',')[2])

print((kusok<dlinna*shirina) and (kusok%shirina==0 or kusok%dlinna==0))
```

```python
"""
Задание 5: Проверка на вещественное число
Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)
"""
any_string = input('Введите строку')
# все элементы которые могу содержаться в натуральном положительном числе
numbers_with_dot = ['1','2','3','4','5','6','7','8','9','0','.']
is_digit = True
for char in any_string:
    if not char in numbers_with_dot:
        is_digit=False
        break
print(is_digit)
```
