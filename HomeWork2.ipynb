```python
from random import random

"""
Написать программу, которая получает на вход строку и возвращает словарь, где:
ключи — символы из этой строки;

значения — количество раз, сколько каждый символ встречается.


"""
input_str = input('Введите строку:')
out_dict = {}

for char in input_str:
    if char in out_dict.keys():
        out_dict[char] += 1
    else:
        out_dict[char] = 1

print(out_dict)
```

```python
"""
Даны три (или больше) списка с объектами. Программа должна создать новый список, содержащий все уникальные элементы — каждый объект встречается только один раз.
"""

input_list_1 = ['dd','aa',876,987]
input_list_1 = ['aa','ll',85,'dd',98,{'kg':25,'if':878}]
input_list_3 = ['ss',[253,'7a8'],987]
combined_list = input_list_1 + input_list_1 + input_list_3
out_lst = []
for char in combined_list:
    if not char in out_lst:
        out_lst.append(char)
print(out_lst)
```

```python
"""
Дана строка текста (или введённая через консоль). Программа должна вернуть словарь с четырьмя ключами:
"гласные",
"согласные",
"цифры",
"пунктуация".
Значения — количество символов каждого типа в строке
"""

input_str = input('Введите строку:')
out_dict = {"vowels":0,"consonants":0,"digit":0,"signs":0}
l_vowels = 'aeiou'
l_signs = ',.:-&@#$%^&!(){}:;><=+?'
for char in input_str:
    if char in l_vowels:
        out_dict['vowels'] += 1
    elif char in l_signs:
        out_dict['signs'] += 1
    elif char.isdigit():
        out_dict['digit'] += 1
    else:
        out_dict['consonants'] += 1
print(out_dict)
```

```python
"""
Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

cnt ввел для страховки что бы цикл не превышал 10 итераций
"""

input_num = input('Ведите число')
cnt=0
while len(input_num)>1 and cnt<10:
    input_num = str(sum([int(i) for i in input_num ]))
    cnt += 1
print(input_num)
```

```python
"""
Задание 2: Кинотеатр
Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

Рандомно сажаю. Потом подбираю ряды под необходимую посадку.
Выводит ряды на которых можно разместиться, если конечно посадка позволит.
Можно все параметы брать и из ввода из консоли
"""

from random import randint
ryad_cnt = int(input('Введите кол-во рядов:')) # кол-во рядов
place_cnt = int(input('Введите кол-во мест в ряду:')) # кол-во мест в ряду
seat_cnt = int(input('Введите кол-во билетов:')) # кол-во билетов необходимых

# рассадка
seats = []
for i in range(0,ryad_cnt):
    etalon = [randint(0,1) for x in range(place_cnt) ]
    seats.append(etalon)

# ищем хоть один ряд куда можно сесть рядом
result = False

for ryad in seats:
     for smehenie in range(0,place_cnt-seat_cnt+1):
         if sum([x for x in ryad[smehenie:smehenie+seat_cnt]])==0:
             print(ryad)
             result=True
             break
     if result:
        break

if not result:
    print(False)
```

```python
"""
Задание 3: Алгоритм RLE
Необходимо написать упрощенную версию алгоритма RLE. Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.
"""

base_str=input('введите строку')
base_str += ' '
new_list = []#резулютирующий список
simbol = base_str[0]#промежуточный символ
for i in range(1, len(base_str)):
    if base_str[i-1] != base_str[i]:# если следующий символ не равен предыдущему
       new_list.append(simbol)#добавляем символ в список
       simbol = base_str[i]#меняем символ
    else:
       simbol += base_str[i]#повторяем символ
print(''.join([str(str(len(i))+i[0]) for i in new_list]))#список символов с количеством

```

```python
"""
Задание 4: Шифр Цезаря
Реализуйте программу для шифрования текста с помощью шифра Цезаря.
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.
"""
alfa_cnt=26# число букв в алфавите
input_str = input('Введите строку')
step=int(input('Введите шаг'))
result_str=''
alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
for char in input_str:
    if char.upper() in alphabet:
        ind_x =  alphabet.index(char.upper(),0,len(alphabet))
        if char.isupper():
            result_str += alphabet[(ind_x+step)%alfa_cnt]#отсток от деления если шаг слишком большой все равно все по кругу
        else:
            result_str += alphabet[(ind_x+step)%alfa_cnt].lower()
    else:
      result_str += char
print(result_str)
```

```python
"""
Задание 5: Табель успеваемости
Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

"""


mark_table = {}
# Цикл ввода данных
while True:
    line = input("Введите данные в формате: название предмета фамилия ученика оценка через пробел\nДля завершения введите пустую строку")
    if line == '':
        break

    parts = line.split()
    subject = parts[0]  # Предмет
    student = parts[1]  # фамилия
    mark = parts[2]    # оценка

    #словарь в словаре и оценки как список надо инициализировать
    if not subject in mark_table:
        mark_table[subject]={}
    if not student in mark_table[subject]:
        mark_table[subject][student]=[]

    mark_table[subject][student].append(mark)



for subject, students in mark_table.items():
    print(subject)
    for student, mark in students.items():
        print(student , ' '.join(mark))
```
