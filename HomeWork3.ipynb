```python
from numpy import character

"""
Написать функцию, которая принимает неограниченное количество чисел в виде позиционных аргументов и ключевой аргумент — операцию над этими числами (сложение или умножение).
Функция должна возвращать результат выполнения указанной операции.
"""
def OnwOperation(operation : str,*args : int)->int:
    if operation=='+':
        result = 0
    if operation=='*':
        result = 1
    for i in args:
        if operation == '+':
            result += i
        elif operation == '*':
            result *= i
        else:
            result=0
    return result

print(OnwOperation('*',1,3,5,7,8,9))#умножение
print(OnwOperation('+',1,3,5,7,8,9))#сложение
print(OnwOperation('/',1,3,5,7,8,9))#ничего
```

```python
"""
Написать функцию, которая создаёт абсолютный путь к файлу.

Позиционные аргументы:

название диска,

неограниченное количество папок,

имя файла (без расширения).


Ключевые аргументы:
ext — расширение файла,

sep — разделитель (по умолчанию '/').
"""


def open_file(drive : str,name : str,*args, sep :str='/',ext : str)->str:
    result = drive+':'+sep
    for path in args:
        result += path+sep
    result += name+'.'+ext
    return result

open_file('c', 'work', 'python', 'function', 'main',ext='py')
```

```python
"""
Написать функцию, которая принимает список, состоящий из объектов разных типов, и возвращает словарь, где:
ключи — типы данных объектов;

значения — списки объектов соответствующего типа.
"""

def define_objects(*args)->dict:
    result = {}
    for in_args in args:
        if not in_args.__class__.__name__ in result:
            result[in_args.__class__.__name__] = []
        result[in_args.__class__.__name__].append(in_args)
    return result

print(define_objects('1','dasd','das',1,[35,68],17.48,('11','dst'),2.58,True,{'fr':54},False))
```

```python
"""
Задание 1: Конвертер регистров
Написать функцию, которая будет переводить snake_case в PascalCase и наоборот.

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.


"""

import re #регулярные выражения надоел цикл

def to_snake(stroka:str):
    ## заменяем последовательность Большая плюс маленькая на тоже самое но с подчеркиванием
    stroka = re.sub('([a-z])([A-Z])', r'\1_\2', stroka)

    return stroka.lower()

def upper(match):
    return match.group(0).upper().replace('_','')

def to_paskal(stroka:str):
    ## заменяем последовательность _ плюс маленькая на тоже самое но с подчеркиванием
    stroka = re.sub('(_)([a-z])',upper, stroka)
    return stroka[0].upper()+stroka[1::1]

def define_and_print(input_str:str)->str:
    if input_str.find('_',1,len(input_str))>0:
        return to_paskal(input_str)
    else:
        return to_snake(input_str)


print(define_and_print('PythonIsTheBestUnit'))
print(define_and_print('otus_course_base'))
```

```python
"""
Написать функцию проверяющую валидность введенной даты.
"""
days_in_month = [31, 29, 31, 30, 31, 30,31, 31, 30, 31, 30, 31] #дней в гду по месяцас начиная с января

def chek_date(date_str:str)->bool:
    result = True
    day, month, year = map(int, date_str.split('.'))
    if not (1 <= day <= 31) or not (1 <= month <= 12): # дней от 1  до 31 и месяцев 1 до 12
        result = False
    if month == 2 and day > 29: # в феврале не более 29 дней
            result = False
    if year % 4 == 0 : #для февраля  высокосный год
        if month == 2 and day > 29:
            result = False
        else:
           if day > days_in_month[month - 1]:
               result = False
    return result

date_str = input('введите дату в формате dd.mm.yyyy:')
print(chek_date(date_str))



```

```python
"""
Проверка на простое число
Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.


"""
def is_simple(number)->bool:
    if number<2:
        return False
    сnt = 0
    for i in range(2, number // 2 +1):#перебераем только половину чисел + 1
        if (number % i == 0):
            сnt += 1
    if (сnt == 0):#нет чисел кроме себя и 1
        return True
    else:
        return False

is_simple_number  = int(input('Введите число'))
print(is_simple(is_simple_number))

```

```python
"""
Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле.

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа.

Программа должна проверять:

имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
возраст должен быть числом от 18 до 60;
ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

Пример
Сергей,петров,23,12
алексей,Пет3ров,3,102
Афанасий,Погорелов,18,102

Если не подходит по критериям то не записываем
"""

def fio(name:str)->bool:
    if not name.isalpha() :
        return False
    return True

def age(number:int)->bool:
    if not number in range(18,60):
        return False
    return True

def IDs(number:int)->bool:
    if not number in range(1,99999999):
        return False
    result = str(number).zfill(8)
    if result in users:
        return False
    return True

users = {}

while True:
    new_user =  input('Введите через запятую пользователя в формате Имя,Фамилия,возраст,ID')
    if new_user=='':
        break
    first_name,second_name,user_age,user_IDs=new_user.split(',')
    if (not fio(first_name)==True) or (not fio(second_name)==True) :
        continue
    if not age(int(user_age))==True :
        continue
    if not IDs(int(user_IDs))==True :
        continue
    users[user_IDs.zfill(8)]=()
    users[user_IDs.zfill(8)]=(first_name,second_name,user_age)

print (users)
```
