```python
import os
from abc import abstractmethod


"""
Базовый класс
"""
class BaseMedia:
    FName : str
    Size : int
    DCreate : float
    OwnerName : str
    Content : []

    def __init__(self,FPath):
        """
        Конструктор читает файл
        """
        self.FName = os.path.basename(FPath)
        self.DCreate = os.stat(FPath).st_atime
        self.Size = os.stat(FPath).st_size
        self.OwnerName = os.stat(FPath).st_uid
        self.Content = []

        with open(FPath, "r") as file:
            self.Content  = file.readline()

    def __repr__(self):
        """
        Вывод на печать
        """
        return f'Файл {self.FName} , размер {self.Size} , создан {self.DCreate} , владелец {self.OwnerName}'



    @abstractmethod
    def PlayMedia(self):
        """
        Воспроизведение медиа обязан быть у каждого наследника
        """
        pass

    def __add__(self, other) -> bite:
        """
        Без реализации просто монтаж медиа
        :param other:
        :return:
        """
        self.Size += other.Size
        self.Content += other.Content
        return self.Content

    def SaveMedia(self):
        """
        Запись в файл
        :return:
        """
        with open(self.FName, "w") as file:
            file.write(self.Content)



```
